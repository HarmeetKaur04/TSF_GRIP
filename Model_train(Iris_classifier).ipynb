```python
import numpy as np
import pandas as pd
```

```python
iris=pd.read_csv("iris.csv")
iris.head()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
iris['Species'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
iris['Species']=le.fit_transform(iris['Species'])
```

```python
iris.head()
```

```python
X=iris.iloc[:,1:-1]
y=iris.iloc[:,5:]
```

```python
X=X.values
```

```python
X
```

```python
y=y.values
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(9, 3.5))

plt.subplot(121)
plt.plot(X[np.where(y==0), 2], X[np.where(y==0), 3], "yo")
plt.plot(X[np.where(y==1), 2], X[np.where(y==1), 3], "bs")
plt.plot(X[np.where(y==2), 2], X[np.where(y==2), 3], "g^")
plt.xlabel("PetalLengthcm")
plt.ylabel("PetalWidthcm")
plt.grid()
plt.title("yellow for setosa \n blue for versicolor \n green for virginica")
plt.subplot(122)
plt.scatter(X[:, 2], X[:, 3], c="k", marker=".")
plt.xlabel("PetalLengthcm")
plt.tick_params(labelleft=False)
plt.gca().set_axisbelow(True)
plt.grid()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans_per_k = [KMeans(n_clusters=k, n_init=10, random_state=42).fit(X)
                for k in range(1, 10)]
inertias = [model.inertia_ for model in kmeans_per_k]

plt.figure(figsize=(8, 3.5))
plt.plot(range(1, 10), inertias, "bo-")
plt.xlabel("$k$")
plt.ylabel("Inertia")
plt.annotate("", xy=(3, inertias[3]), xytext=(4.45, 650),
             arrowprops=dict(facecolor='black', shrink=0.1))
plt.text(4.5, 650, "Elbow", horizontalalignment="center")
plt.axis([1, 8.5, 0, 1300])
plt.grid()
plt.show()
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
# Hence the optimum clusters would be 3 because using elbow method the slight change in graph is on k=3.

thank you.
```
